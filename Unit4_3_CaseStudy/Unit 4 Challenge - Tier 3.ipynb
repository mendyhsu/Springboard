```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
```

#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file.
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.

properties = pd.read_excel(url_LondonHousePrices,
                           sheet_name='Average price',
                           index_col=None,
                           header=None)
# [Mendy's note] adding header=None to not assign the column names
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
propertiesT.iloc[0,0:2]=['Borough','Zip']
propertiesT.head()
```

```python
col_names = propertiesT.iloc[0] # first row for the header
d_prop = propertiesT[1:]        # remove the header row from data
d_prop.columns = col_names     # set the header row as column name
d_prop.head()
```

```python
display(d_prop.shape)
d_prop.describe()
#d_prop.dtypes

# [Mendy's note] data has 48 rows, but only 45 of them have values/conts
```

```python
# Check any rows that contain "any NaN"
d_prop[d_prop.isna().any(axis=1)]
```

```python
# Drop the 3 rows with NaN
row_to_drop = d_prop.isna().any(axis=1)
d_prop = d_prop[~row_to_drop]
d_prop.shape

# [Mendy's note] Number of rows = 45 that is more than the Number of Borough (32).
```

```python
# From Wiki (https://en.wikipedia.org/wiki/List_of_London_boroughs)
LDN_Borough_wiki = pd.read_excel('London_Borough_list.xlsx',header=0) 
LDN_Borough_list = LDN_Borough_wiki['Borough']
LDN_Borough_list
```

```python
# Checking if the name of the Boroughs (d_prop['Borough']) overlaped with those in LDN_Borough_list
Borough_interset = set(d_prop['Borough'].values).intersection(LDN_Borough_list.values)
len(Borough_interset)
```

```python
# Selecting the rows that is within the "32 Boroughs"
LDN_Borough_prop = d_prop[d_prop['Borough'].isin(Borough_interset)]
LDN_Borough_prop.shape
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
df = pd.melt(LDN_Borough_prop,
             id_vars=['Borough', 'Zip'],
             var_name='Month',
             value_name='Average_Price')
display(df.info())
df.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Change the dtypte of 'Average_Price'
df['Average_Price']=df['Average_Price'].astype('float')
```

```python
display(df.info())
display(df.head())
#display(df.describe())
df.dtypes
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Random draw 5 LND Boroughs, and show their "Avg. Price VS Time"
rand_draw = np.random.choice(list(Borough_interset), 5)

fig, ax = plt.subplots(1, 1, figsize=(15, 10), dpi=60)

for i, borough_name in enumerate(rand_draw):
    df[df['Borough'] == borough_name].plot(kind='line',
                                           x='Month',
                                           y='Average_Price',
                                           ax=ax)

ax.set_xlabel('Year', fontsize=20)
ax.set_ylabel('Average price (GBP)', fontsize=20)
ax.legend(rand_draw, fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.tick_params(axis='both', which='minor', labelsize=16)
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
df['Year'] = df['Month'].apply(lambda x: x.year)
df.head()
```

```python
df_yearly = df.groupby(['Borough','Zip','Year']).mean()
df_yearly = df_yearly.reset_index('Year')
df_yearly.head()
```

```python
df_yearly.loc[(list(Borough_interset)[0], slice(None)),:].plot(kind='line', x='Year', y='Average_Price')
plt.title(list(Borough_interset)[0])
plt.show()
```

```python
# Unset index
df_yearly_reset = df_yearly.reset_index()[['Borough', 'Year', 'Average_Price']]
df_yearly_reset.head()
```

```python
# Make a pivot table, Borough (row) by Year (col)
df_borough_by_year = df_yearly_reset.pivot_table(index=['Borough'],
                                                 columns='Year',
                                                 values='Average_Price')
df_borough_by_year.head()
```

```python
# Summary of housing price by year, across 32 LND boroughs
df_borough_by_year.describe()
```

```python
# The range of prices
fig = plt.figure(figsize=(12, 5), dpi=80)
(0.001*df_borough_by_year).boxplot()
plt.title('Range of housing prices from 1998 to 2018 ', fontsize=16)
plt.ylabel('Housing prices (1K GBP)', fontsize=16)
plt.xticks(rotation=90, fontsize=16)
plt.show()
```

```python
# Compute the "percentage changes" in Avg_Price relative to year 1998
price_1998 = df_borough_by_year.loc[:, 1998]

price_change = 100 * df_borough_by_year.sub(price_1998, axis='rows').div(
    price_1998, axis='rows')

price_change.head()
```

```python
fig = plt.figure(figsize=(12, 5), dpi=80)

price_change[2018].sort_values(ascending=False).plot(kind='bar')

plt.title('% Change in housing price, bwtween 1998 and 2018', fontsize=15)
plt.xlabel('')
plt.ylabel('% changes btw. yearly prices', fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.show()
```

```python
price_change[2018].sort_values(ascending=False)
```

```python
sort_by_price_change = list(price_change[2018].sort_values(ascending=False).index)
df_borough_by_year.T.loc[1998:2018, sort_by_price_change].describe()
```

```python
# The housing price of each borough averaged over years 
df_borough_by_year.T.mean().sort_values(ascending=False)
```

```python
fig = plt.figure(figsize=(12, 5), dpi=80)
(0.001*df_borough_by_year).T.loc[1998:2018,sort_by_price_change].boxplot()
plt.title('Housing prices range of each borough (from 1998 to 2018)', fontsize=16)
plt.ylabel('Housing prices (1K GBP)', fontsize=16)
plt.xticks(rotation=90, fontsize=16)
plt.show()
```

```python
# Prepaering the table for showing "Percentage changes in price VS Time"
price_change_unind = price_change.reset_index()
price_change_unind.head()
```

```python
#price_change.T.describe()
```

```python
top_ten = sort_by_price_change[:10]

fig, ax = plt.subplots(1, 1, figsize=(15, 10), dpi=60)

price_change.T[top_ten].plot(kind='line', ax=ax) # Show top 10 boroughs

# The change in "mean price" (avg. over boroughs) VS year
(100 * (df_borough_by_year.mean() - df_borough_by_year.mean()[1998]) /
 df_borough_by_year.mean()[1998]).plot(kind='line', style='k-.', ax=ax)

ax.set_title('Top 10 London boroughs with biggest changes in housing prices', fontsize=25)
ax.set_xlabel('Year', fontsize=22)
ax.set_ylabel('% change in price (relative to 1998)', fontsize=22)
ax.legend(top_ten + ['Mean'], fontsize=16, bbox_to_anchor=(1, 1))
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='minor', labelsize=20)
plt.show()
```

```python
# Alternative way to make the same plots
df_price_change = pd.melt(price_change_unind,
                          id_vars=['Borough'],
                          var_name='Year',
                          value_name='Average_Price')

df_price_change = df_price_change.set_index('Borough')
display(df_price_change.info())
df_price_change.head()
```

```python
# Top 10 LND Boroughs: "Percentage changes in price VS Time" relative to 1998
top_ten = sort_by_price_change[:10]

fig, ax = plt.subplots(1, 1, figsize=(15, 10), dpi=55)

for i, borough_name in enumerate(rand_draw):
    df_price_change.loc[(borough_name), :].plot(kind='line',
                                                x='Year',
                                                y='Average_Price',
                                                ax=ax)

ax.set_title('Changes housing prices of 10 London boroughs', fontsize=25)
ax.set_xlabel('Year', fontsize=22)
ax.set_ylabel('% change in price (relative to 1998)', fontsize=22)
ax.legend(top_ten, fontsize=16, bbox_to_anchor=(1, 1))
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='minor', labelsize=20)
plt.show()
```
